```python
# import dependencies 
import pandas as pd
from pathlib import Path

#load data from resources folder
school_data_csv = Path("Resources/schools_complete.csv")
student_data_csv = Path("Resources/students_complete.csv")

# Read School and Student Data and save as Panda DataFrames
school_data = pd.read_csv(school_data_csv)
student_data = pd.read_csv(student_data_csv)

#combine data into single dataframe

school_data_completed = pd.merge(student_data,school_data, how="left", on=["school_name","school_name"])
school_data_completed.head()
```

```python
school_value_count = school_data_completed["school_name"].value_counts()
school_count = school_value_count.count()
school_count
```

```python
student_count = school_data_completed["student_name"].count()
student_count
```

```python
total_budget = school_data_completed["budget"].unique().sum()
total_budget
```

```python
average_math_score = school_data_completed["math_score"].mean()
average_math_score
```

```python
average_reading_score = school_data_completed["reading_score"].mean()
average_reading_score
```

```python
passing_math_count = school_data_completed[(school_data_completed["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
passing_reading_count = school_data_completed[(school_data_completed["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_completed[
    (school_data_completed["math_score"] >= 70) & (school_data_completed["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
district_summary = pd.DataFrame({
    "Total Schools": school_count,
    "Total Students": student_count,
    "Total Budget": total_budget,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": passing_math_percentage,
    "% Passing Reading": passing_reading_percentage,
    "% Overall Passing": overall_passing_rate
})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary = district_summary.reset_index(drop=True)
district_summary.head(1)
```

```python
per_school_groups = school_data_completed.groupby("school_name")
school_types = per_school_groups["type"].unique()
school_types
```

```python

per_school_counts = per_school_groups["Student ID"].count()
per_school_counts
```

```python
per_school_budget = per_school_groups["budget"].unique()
per_school_budget = per_school_budget.astype({"budget":int})
per_school_budget

per_school_capita = per_school_budget/per_school_counts
per_school_capita
```

```python
per_school_math = per_school_groups["math_score"].mean()
per_school_math
per_school_reading = per_school_groups["reading_score"].mean()
per_school_reading
```

```python
students_passing_math = school_data_completed[(school_data_completed["math_score"]>=70)]
per_school_passing_math = students_passing_math.groupby(["school_name"]).size()
per_school_passing_math

students_passing_reading = school_data_completed[(school_data_completed["reading_score"]>=70)]
per_school_passing_reading = students_passing_reading.groupby(["school_name"]).size()
per_school_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_completed[
    (school_data_completed["reading_score"] >= 70) & (school_data_completed["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading
```

```python
percent_school_passing_math = per_school_passing_math / per_school_counts * 100
percent_school_passing_reading = per_school_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100

percent_school_passing_math
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": per_school_counts,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_school_capita,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "% Passing Math": percent_school_passing_math,
    "% Passing Reading": percent_school_passing_reading,
    "% Overall Passing": overall_passing_rate
})
# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

# Display the DataFrame
per_school_summary
```

```python
top_schools = per_school_summary.sort_values("% Overall Passing", ascending=False)
top_schools.head()
```

```python
bottom_schools = per_school_summary.sort_values("% Overall Passing", ascending=True)
bottom_schools.head()
```

```python
ninth_graders = school_data_completed[(school_data_completed["grade"] == "9th")].groupby("school_name")
tenth_graders = school_data_completed[(school_data_completed["grade"] == "10th")]
eleventh_graders = school_data_completed[(school_data_completed["grade"] == "11th")]
twelfth_graders = school_data_completed[(school_data_completed["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders["math_score"].mean()
tenth_grader_math_scores = tenth_graders["math_score"].mean()
eleventh_grader_math_scores = eleventh_graders["math_score"].groupby("school_name").mean()
twelfth_grader_math_scores = twelfth_graders["math_score"].groupby("school_name").mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({
    "9th": ninth_grade_math_scores,
    "10th": tenth_grader_math_scores,
    "11th": eleventh_grader_math_scores,
    "12th": twelfth_grader_math_scores,
})
# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```
